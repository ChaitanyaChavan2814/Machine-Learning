```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
target = df.Survived
inputs = df.drop('Survived',axis='columns')
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs.drop('Sex',axis='columns',inplace=True)
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
df1 = pd.read_csv('spam.csv')
df1.head()
```

```python
df1.groupby('Category').describe()
```

```python
df1['spam'] = df1['Category'].apply(lambda x: 1 if x=='spam' else 0)
df1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1.Message,df1.spam,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#convert a collection of text documents to a matrix of token counts.
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

fit_transform is called on the CountVectorizer instance v with the training data X_train.values.

fit learns the vocabulary dictionary of all tokens in the raw documents.

transform converts the text data to a matrix of token counts.

X_train_count now contains the token counts for the training data in matrix form.

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

The transform method is called on the CountVectorizer instance v with the new emails.

This converts the new emails to a matrix of token counts using the vocabulary learned during the fitting step.

# sklearn pipeline

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
