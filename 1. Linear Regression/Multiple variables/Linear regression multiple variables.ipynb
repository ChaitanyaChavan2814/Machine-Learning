```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices.csv')
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())  #round down the median value
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,3,40]])
```
