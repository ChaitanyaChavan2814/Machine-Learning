```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area(sq ft)')
plt.ylabel('Price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
X = df[['area']]  # Features
# treat input variable in 2d format
y = df['price']   # Target variable

# Creating an instance of LinearRegression
reg = linear_model.LinearRegression()

# Fitting the model
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[5000]])
```

```python
# y=mx+b to predict price for 5000 sq ft
135.78767123*5000+180616.43835616432
```

```python
df1=pd.read_csv('areas.csv')
df1.head(3)
```

```python
p = reg.predict(df1[['area']])
df1['prices']=p
df1
```

```python
df1.to_csv("prediction.csv",index='False')
```

```python
%matplotlib inline
plt.xlabel('Area(sq ft)',fontsize=20)
plt.ylabel('Price(US$)',fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
