```python
import pandas as pd
df = pd.read_csv('homeprices.csv')
df
```

```python
dummies = pd.get_dummies(df.town, dtype=int)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
final = merged.drop(['town','west windsor'],axis='columns')
final
# rule of thumb: always drop one dummy variable column
```

```python
from sklearn.linear_model import LinearRegression
model =  LinearRegression()
```

```python
X = final.drop('price',axis='columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
# To know how accurate model is
model.score(X,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
X
```

```python
y = dfle.price
y
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(), [0])  # Encode column at index 0
    ],
    remainder='passthrough'  # Keep remaining columns unchanged
)

X = ct.fit_transform(X)
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[0,1,3400]])
```
